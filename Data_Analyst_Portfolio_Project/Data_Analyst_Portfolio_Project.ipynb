```python
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
df2.head(10)
```

```python
 #remove (USA) from Event name
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
#clean up athlete age
df2['Athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#remove h from athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.replace('h','')
```

```python
df2.head(10)
```

```python
#drop columns : athlete club, athlete country, athlete year of birth, athlete age category
df2 = df2.drop(columns=['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], errors='ignore')
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#check for duplicates
df2.duplicated().sum()
```

```python
#reset index
df2 = df2.reset_index(drop=True)
```

```python
df2['Athlete_age'] = df2['Athlete_age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
#rename columns
# Year of event                  int64
# Event dates                   object
# Event name                    object
# Event distance/length         object
# Event number of finishers      int64
# Athlete performance           object
# Athlete gender                object
# Athlete average speed        float64
# Athlete ID                     int64
# Athlete_age                    int32
df2 = df2.rename(columns={
    'Year of event': 'year',
    'Event dates': 'race_date',
    'Event name': 'race_name',
    'Event distance/length': 'race_length',
    'Event number of finishers': 'race_number_of_finishers',
    'Athlete performance': 'athlete_performance',
    'Athlete gender' : 'athlete_gender',
    'Athlete average speed': 'athlete_average_speed',
    'Athlete ID': 'athlete_id',
    'Athlete_age': 'athlete_age'
    
    
    
    })
```

```python
#Different in speed for the 50k, 50mi male to female
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
#what age group are the best in the 50m Race (20 + race min)
df3.query('race_length == "50mi" ').groupby('athlete_age')['athlete_average_speed'].agg(["mean", "count"]).sort_values(by='mean', ascending=False).query('count > 19')
```

```python
#what age group are the worst in the 50m Race (10 + race min) (show 20)
df3.query('race_length == "50mi" ').groupby('athlete_age')['athlete_average_speed'].agg(["mean", "count"]).sort_values(by='mean', ascending=True).query('count > 9')
```

```python
#season for the data -> slower in summer than winter?

#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2

#split between the two decimal points
```

```python
df3['race_month'] = df3['race_date'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'spring' if x in [3,4,5] else ('summer' if x in [6,7,8] else ('fall' if x in [9,10,11] else 'winter')))
```

```python
df3.head(10)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values(by='mean', ascending=False)
```

```python
#50 miler only
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values(by='mean', ascending=False)
```

```python
import matplotlib.pyplot as plt

df3.query('race_length == "50mi"') \
   .groupby('race_season')['athlete_average_speed'] \
   .mean().reindex(['winter', 'spring', 'summer', 'fall']).plot(kind='bar', color='skyblue')

plt.ylabel('Average Speed (mph)')
plt.title('Average Speed by Season (50mi races)')
plt.show()
```
